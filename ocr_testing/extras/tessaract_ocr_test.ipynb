```python
import pytesseract
from PIL import Image

# Path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'E:\Projects\tessaract\tesseract.exe'

# Path to the image file
image_path = r'E:\Projects\OCR\ocr_testing\samples\gray_images\gray_Screenshot_20250211-234427_1.png'

# Open the image file
image = Image.open(image_path)

def extract_text_pytesseract(image):
    """
    Extracts text, bounding box, and confidence from an image using pytesseract.

    Args:
        image: The input image.

    Returns:
        A list of dictionaries containing extracted text, bounding boxes, and confidence scores.
    """
    extracted_data_lst = []

    # Perform OCR with pytesseract (gives detailed structured data)
    data = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)

    for i in range(len(data['text'])):
        text = data['text'][i].strip()
        confidence = float(data['conf'][i]) if data['conf'][i] != '-1' else 0.0

        if text:  # Ignore empty results
            item = {
                'bounding_box': [
                    [data['left'][i], data['top'][i]],
                    [data['left'][i] + data['width'][i], data['top'][i]],
                    [data['left'][i] + data['width'][i], data['top'][i] + data['height'][i]],
                    [data['left'][i], data['top'][i] + data['height'][i]],
                ],
                'text': text,
                'confidence': confidence / 100  # Convert to 0-1 scale
            }
            extracted_data_lst.append(item)

    return extracted_data_lst

# Extract text from the image
texts = extract_text_pytesseract(image)

# Print the extracted data
print(texts)
```
